# Central limit theorem

There are many algorithms and statistics that require the assumption of normality, such as hypothesis testing, confidence intervals, and other statistical methods. We may have a dataset that is not normally distributed, but we can still use the central limit theorem to make inferences about the population mean. 

The central limit theorem states that, given a sufficiently large sample size, the distribution of the sample mean (or sum) will be approximately normally distributed, regardless of the shape of the original population distribution. This allows us to make inferences about the population mean of any distribution, as long as the sample size is large enough.

## Data

Let's create some datasets and visualize them.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

np.random.seed(42)
exponential_data = np.random.exponential(scale=2, size=10000)  # Exponential distribution
uniform_data = np.random.uniform(low=0, high=1, size=10000)  # Uniform distribution

sns.histplot(exponential_data, kde=True, label='Exponential Data', bins=30)
plt.title('Histogram of Exponential Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

sns.histplot(uniform_data, kde=True, label='Uniform Data', bins=30)
plt.title('Histogram of Uniform Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## Compute the means

Let's compute the sample means for the datasets and visualize the distribution of the sample means.

```python
# Draw multiple samples and compute their means
sample_size = 1000  # Sample size
n_samples = 500  # Number of samples

exponential_means = [np.mean(np.random.choice(exponential_data, size=sample_size)) for _ in range(n_samples)]
uniform_means = [np.mean(np.random.choice(uniform_data, size=sample_size)) for _ in range(n_samples)]

sns.histplot(exponential_means, kde=True, label='Exponential Means', bins=30)
plt.title('Histogram of Exponential Means')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

sns.histplot(uniform_means, kde=True, label='Uniform Means', bins=30)
plt.title('Histogram of Uniform Means')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## Normality test

Let's compute the Shapiro-Wilk test to check if the sample means are normally distributed.

```python
# Shapiro-Wilk test for normality
from scipy.stats import shapiro, kstest
# Exponential data
stat, p = shapiro(exponential_means)
print(f'Shapiro-Wilk test for Exponential Means: p-value = {p:.4f}')
print(f"Is the sample mean of the exponential data normally distributed? {p > 0.05}")
stat, p = shapiro(uniform_means)
print(f'Shapiro-Wilk test for Uniform Means: p-value = {p:.4f}')
print(f"Is the sample mean of the uniform data normally distributed? {p > 0.05}")
```
